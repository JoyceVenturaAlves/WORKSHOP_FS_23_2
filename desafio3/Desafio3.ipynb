#nota média em Matemática a partir da raça/etnia dos alunos

```python
df2 = df.groupby('math score')[['race/ethnicity']].count()
df2
```

```python
df2.plot(kind= 'bar')
```

#nota média em Escrita a partir do Nível de Escolaridade dos pais

```python
df3 = df.groupby('writing score')[['parental level of education']].count()
df3
```

```python
df3.plot(kind= 'bar')
```

#nota média em Leitura a partir da partificipação em curso preparatório para os testes

```python
df4 = df.groupby('reading score')[['test preparation course']].count()
df4
```

```python
df4.plot (kind= 'bar')
```
