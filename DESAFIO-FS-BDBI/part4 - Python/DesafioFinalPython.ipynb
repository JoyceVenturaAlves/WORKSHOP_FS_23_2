```python
# Visualizar as primeiras linhas do banco de dados
data.head()
```

```python
# 1. Características do Banco de Dados

# a. Quantas linhas e colunas temos no banco de dados?
num_rows, num_columns = data.shape
print(f"Número de linhas: {num_rows}")
print(f"Número de colunas: {num_columns}")

# b. Existem dados nulos no banco? Se sim, quais colunas apresentam e quantos são os dados nulos nelas?
null_values = data.isnull().sum()
print("Dados nulos por coluna:")
print(null_values)

# c. Quais tipos de dados estão presentes no banco de dados?
data_types = data.dtypes
print("Tipos de dados por coluna:")
print(data_types)

# 2. Agrupamento de Dados e Plotagem de Gráficos

# a. Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department);
data['Department'].value_counts().plot(kind='bar')
plt.title('Quantidade de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Quantidade de Funcionários')
plt.xticks(rotation=45)
plt.show()

# b. Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education);
data['Education'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Quantidade de Funcionários por Nível de Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Funcionários')
plt.xticks(rotation=45)
plt.show()

# c. Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que têm ou não algum atrito na empresa (coluna Attrition);
attrition_counts = data['Attrition'].value_counts()
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentual de Funcionários com Atrito na Empresa')
plt.axis('equal')
plt.show()
```
